```python
from linkedin_scraper import Person,actions
from selenium import webdriver
import os
```

```python
email='rafaelgov95@gmail.com'
senha=os.path['senha_']
driver =webdriver.Chrome()
actions.login(driver, email,password=senha,timeout=1000)
```

```python
import time
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
save_link = []
driver.get("https://ca.linkedin.com/company/ufms/people")
body_locator = (By.TAG_NAME, "body")
WebDriverWait(driver, 10).until(EC.presence_of_element_located(body_locator))
section_locator = (By.CLASS_NAME, "artdeco-card")
section = WebDriverWait(driver, 10).until(EC.visibility_of_element_located(section_locator))
li_locator = (By.CSS_SELECTOR, "li.grid.grid__col--lg-8.block.pt5.pr4.m0")

while True:
    previous_count = len(save_link)
    lis = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located(li_locator))

    for i in range(previous_count, len(lis)):
        try:
            li = lis[i]
            link_element = li.find_element(By.TAG_NAME, "a")
            link = link_element.get_attribute("href")
            save_link.append(link)
            print(f"ID:{i+1} -- {link}")
        except:
            pass

    if len(save_link) == previous_count:
        break

    driver.execute_script("window.scrollTo(0, Math.ceil(document.body.scrollHeight));")
    time.sleep(4)
```
